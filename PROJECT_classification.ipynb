```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df['TotalCharges']=pd.to_numeric(df["TotalCharges"] ,errors='coerce')
df.dropna(subset=['TotalCharges'],inplace=True)
```

```python
df.info()
```

```python
x=df.drop(["Churn","customerID"], axis=1)
y=df["Churn"]
```

```python
x
```

```python
y
```

```python
#feature encoding----> Dummy encoding
x=pd.get_dummies(x,drop_first=True)

x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#DT-CLASSIFIER
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
accuracy_score(y_test,y_pred_dt)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_dt))
```

```python
#applyning the oversampling technique from imblearn randomoversampling
from imblearn.over_sampling import RandomOverSampler
oversample=RandomOverSampler()
#never do oversampling on the the original data if you actually
#want the good preditions
#and results
x_over,y_over=oversample.fit_resample(x_train,y_train)
dt2=DecisionTreeClassifier()
dt2.fit(x_over,y_over)
y_pred_dt2=dt2.predict(x_test)
accuracy_score(y_test,y_pred_dt2)
```

```python
print(classification_report(y_test,y_pred_dt2))
```

```python
#applyning the oversampling technique from imblearn randomoversampling
from imblearn.under_sampling import RandomUnderSampler as Undersample
Downsample=RandomUnderSampler()
#never do oversampling on the the original data if you actually
#want the good preditions
#and results
x_Down,y_Down=Downsample.fit_resample(x_train,y_train)
dt3=DecisionTreeClassifier()
dt3.fit(x_Down,y_Down)
y_pred_dt3=dt3.predict(x_test)
accuracy_score(y_test,y_pred_dt3)
```

```python
print(classification_report(y_test,y_pred_dt3))
```
